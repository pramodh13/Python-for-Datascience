# This module is for extensive EDA and preprocessing analysis -Zomato Dataset

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Import dataset
df=pd.read_csv('zomato.csv',encoding='latin-1')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe() #Only integer features are shown
```

```python
#Finding missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_country=pd.read_excel("Country-Code.xlsx")
```

```python
df_country
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
country_names=final_df["Country"].value_counts().index
```

```python
country_val=final_df["Country"].value_counts().values
```

```python
#Pie chart-Top 3 countries that use zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct="'%1.2f%%'")
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','Red','orange','yellow','green','DarkGreen'])
plt.show()
```

```python
###Find countries name that has given 0 rating
final_df.groupby(["Aggregate rating","Country"]).size().head()
```
