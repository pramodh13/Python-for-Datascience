# Matplotlib libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Initializing data
np.random.seed(12)
df=pd.DataFrame(data={'a':np.random.randint(0,100,30),
                      'b':np.random.randint(0,100,30),
                      'c':np.random.randint(0,100,30)})
df.head()
#df.info()
```

```python
fig=plt.figure(figsize=(5,8))
%matplotlib inline
ax=plt.subplot(1,1,1)
```

```python
#Subplots unpacking
fig,ax=plt.subplots(1,3,figsize=(15,2))
ax[0].plot(df.index.values,df['a'])
ax[1].plot(df.index.values,df['b'])
ax[2].plot(df.index.values,df['c'])
```

```python
#Line plots with multiple axis and all features
sns.set_style('darkgrid')
fig,ax=plt.subplots(1,1,figsize=(10,4))
ax.plot(df.index.values,df['a'],color='red',ls='-.',label='Line A')
ax.plot(df.index.values,df['b'],color='blue',lw=2,label='Line B')
ax.plot(df.index.values,df['c'],color='yellow',lw=1,marker='o',label='Line C')
ax.legend(loc='best')
```

```python
#Barplots and its params
fig,ax=plt.subplots(1,1,figsize=(5,5))
bar_kwargs={'color':'tomato','alpha':0.5}
ax.bar(df.index.values,df['a'],label='a',**bar_kwargs)
ax.legend()
```

```python
#Scatter plot
plt.plot([1,2,3,4,5],[2,4,6,8,10],'go',label='Green dots')
plt.plot([1,2,3,4,5],[10,12,14,2,5],'r*',label='Red star')
plt.title('A Simple Scatterplot')
plt.legend()
plt.show()
```
